# Arbre

**Définition.**
Un arbre est un graphe connexe (entre deux points il existe toujours au moins un chemin) sans boucle (il n'y a pas de chemin dont le point de départ est aussi l'arrivée).

**Objectifs.** 
Comprendre les deux types d'arbres
1. xml et en particulier html
2. python et les ast

# XML

**Exemple.**

```xml
<a>
    <b></b>
    <c>
        <d></d>
        <e></e>
    </c>
</a>   
```

Représente le graphe:
- dont les sommets sont `{'a', 'b', 'c', 'd', 'e'}`
- les arrêtes sont `{('a', 'b'), ('b', 'a'), ('a', 'c'), ('c', 'a'), ('c', 'd'), ('d', 'c'), ('c', 'e'), ('e', 'c')}`
